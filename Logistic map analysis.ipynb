# Logistic Map analysis
In this notebook the logistic map will be studied and shown as an example of a chaotic system.

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams.update({
    'font.family': 'serif',
})
```

**Definition**

The logistic map is defined as:
$$ x_{n+1} = rx_n(1-x_n)$$

```python
def logistic(r, x):
    return r*x*(1-x)

def logistic_k(r, x, k):
    return r*x*(1-x/k)

def exp(r,x):
    return r*x
```

```python
def run_logistic(r, xo, n_vals):
    X = np.zeros(n_vals)
    X[0] = xo

    for i in range(n_vals-1):
        X[i+1] = logistic(r,X[i])
    
    return X
```

**Function analysis**

It resembles the logistic function...

```python
xo = 0.001
n_vals = 100

X = run_logistic(1.1, xo, n_vals)
    
plt.title(f'Logistic function with r=1.1')
plt.xlabel('r')
plt.ylabel('f(x)')
plt.grid()
plt.plot(X)
plt.show()
```

Plotting for various r values

```python
xo = 0.001
n_vals = 200

for r in [1.1, 1.2, 1.4, 1.7]:
    plt.plot(run_logistic(r, xo, n_vals), label=f"r={r}")

plt.title(f'Logistic function with various r values')
plt.xlabel('r')
plt.ylabel('f(x)')
plt.grid()
plt.legend()
plt.show()
```

What is the equilibrium value? 

```python
xo = 0.001
n_vals = 200

r_vals = np.linspace(0.2,2.9,300)
sat_vals = list()

for r in r_vals:
    sat_vals.append(run_logistic(r, xo, n_vals)[-1])

plt.title(f'Saturation values of the logistic function depending on r')
plt.grid()
plt.plot(r_vals, sat_vals)
plt.show()
```

**Finding chaos in a deterministic function**

What happens if you go over 3.0 for the r parameter?

```python
xo = 0.001
n_vals = 100

fig, axs = plt.subplots(3, 3, figsize=(12,9), constrained_layout=True)
r = np.linspace(2.9, 4.0, 9)
for i in range(9):
    ax = axs[i//3][i%3]
    ax.plot(run_logistic(r[i], xo, n_vals))
    ax.set_title(f"Recurrence with r={r[i]}")
    ax.set_xlabel('r')
    ax.set_ylabel('f(x)')

plt.show()
```

**Bifurcation diagram**

Plot the of the last n values for a range of r values

```python
iterations = 1000
last = 100
n = 5000
r = np.linspace(2.5, 4, n)
x = 1e-5 * np.ones(n)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(7,7))
for i in range(iterations):
    x = logistic(r, x)
    if i >= (iterations - last):
        ax.plot(r, x, ',k',alpha=.1)

ax.set_xlim(2.5, 4)
ax.set_title("Bifurcation diagram")
plt.xlabel('r')
plt.ylabel('f(x)')
plt.show()
```
