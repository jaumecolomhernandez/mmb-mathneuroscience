```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def logistic(r, x):
    return r*x*(1-x)

def logistic_k(r, x, k):
    return r*x*(1-x/k)

def exp(r,x):
    return r*x
```

Recurence relation...

```python
xo = 0.001
n_vals = 100

X = np.zeros(n_vals)
X[0] = xo

for i in range(n_vals-1):
    X[i+1] = logistic(1.1,X[i])
    
plt.plot(X)
plt.show()
```

```python
def run_logistic(r, xo, n_vals):
    X = np.zeros(n_vals)
    X[0] = xo

    for i in range(n_vals-1):
        X[i+1] = logistic(r,X[i])
    
    return X
```

Plotting for various r values

```python
xo = 0.001
n_vals = 200

for r in [1.1, 1.2, 1.4, 1.7]:
    plt.plot(run_logistic(r, xo, n_vals), label=f"r={r}")

plt.legend()
plt.show()
```

What is the equilibrium value? 

```python
xo = 0.001
n_vals = 200

r_vals = np.linspace(0.2,2.9,300)
sat_vals = list()

for r in r_vals:
    sat_vals.append(run_logistic(r, xo, n_vals)[-1])

plt.plot(r_vals, sat_vals)
plt.show()
```

Finding chaos in a deterministic function

```python
xo = 0.001
n_vals = 100

fig, axs = plt.subplots(3, 3, figsize=(12,9), constrained_layout=True)
r = np.linspace(2.9, 4.0, 9)
for i in range(9):
    ax = axs[i//3][i%3]
    ax.plot(run_logistic(r[i], xo, n_vals))
    ax.set_title(f"Recurrence with r={r[i]}")

plt.show()
```

Bifurcation diagram

```python
iterations = 1000
last = 100
n = 5000
r = np.linspace(2.5, 4, n)
x = 1e-5 * np.ones(n)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(7,7))
for i in range(iterations):
    x = logistic(r, x)
    if i >= (iterations - last):
        
        ax.plot(r, x, ',k',alpha=.1)
ax.set_xlim(2.5, 4)
ax.set_title("Bifurcation diagram")
plt.show()
```
