# XPPAUT data plots

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.lines import Line2D

matplotlib.rcParams.update({
    'font.family': 'serif',
})
```

```python
results1 = np.loadtxt("/mnt/c/xppall/diagram.DAT")
```

```python
#plt.scatter(results[:,0], results[:,1])
fig, ax = plt.subplots(1,1,figsize=(7, 3.5))

colours = np.array(['b', 'r', 'y', 'g','b'])
plt.scatter(results1[:,0], results1[:,1],s=25,c=colours[results1[:,3].astype(int)],marker="1")
plt.scatter(results1[:,0], results1[:,2],s=25,c=colours[results1[:,3].astype(int)],marker="1")

plt.axvline(x=87, lw=1.2, c="black",linestyle=(0, (3, 5, 1, 5)))
plt.axvline(x=217, lw=1.2, c="black",linestyle=(0, (3, 5, 1, 5)))



npzd_legend = [Line2D([0], [0],marker="s", markerfacecolor='r', markersize=10, color='w', lw=4),
               Line2D([0], [0],marker="s", markerfacecolor='y', markersize=10, color='w', lw=4),
               Line2D([0], [0],marker="s", markerfacecolor='g', markersize=10, color='w', lw=4),
               Line2D([0], [0],marker="s", markerfacecolor='b',  markersize=10, color='w', lw=4)]
plt.legend(npzd_legend, ['Stable node', '?', 'Stable focus','Limit unstable'], loc=2, prop={'size': 10})

# Labels
plt.title('Bifurcation diagram')
plt.xlabel('Iapp (mA)')
plt.ylabel('V (mv)')
plt.grid()

plt.savefig('./img/bifurcation_diagram.pdf')
```

```python
results2 = np.loadtxt("/mnt/c/xppall/diagram_freq.DAT")
```

```python
results2 = results2[np.where(results2[:,3]!=[1])]
results2 = results2[np.where(results2[:,3]!=[2])]
```

```python
#plt.scatter(results[:,0], results[:,1])
fig, ax = plt.subplots(1,1,figsize=(6, 3))

colours = np.array(['k', 'r', 'y', 'g','b'])
plt.scatter(results2[:,0], results2[:,1]*1000,s=15,c=colours[results2[:,3].astype(int)],marker="o")
plt.scatter(results2[:,0], results2[:,2]*1000,s=15,c=colours[results2[:,3].astype(int)],marker="o")

plt.axvline(x=87, lw=1.2, c="black",linestyle=(0, (3, 5, 1, 5)))
plt.axvline(x=217, lw=1.2, c="black",linestyle=(0, (3, 5, 1, 5)))



npzd_legend = [Line2D([0], [0],marker="s", markerfacecolor='g', markersize=10, color='w', lw=4),
               Line2D([0], [0],marker="s", markerfacecolor='b',  markersize=10, color='w', lw=4)]
plt.legend(npzd_legend, ['Periodic','Non Periodic'], loc=2, prop={'size': 10})

plt.ylim([5,25])
plt.xlim([50,250])

# Labels
plt.title('Frequency - Intensity plot')
plt.xlabel('Iapp (mA)')
plt.ylabel('f (Hz)')
plt.grid()
plt.tight_layout()


plt.savefig('./img/fi_plot.pdf')
```
